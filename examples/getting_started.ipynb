## 2. Configure your API key

```python
import os

# Paste values from Veritrix → Get Started
VERITRIX_API_KEY = ""  # @param {type:"string"}
VERITRIX_ENDPOINT = "http://localhost:8001"  # @param {type:"string"}

if not VERITRIX_API_KEY.strip():
    raise ValueError("Set VERITRIX_API_KEY before continuing.")

os.environ["VERITRIX_API_KEY"] = VERITRIX_API_KEY
os.environ["VERITRIX_ENDPOINT"] = VERITRIX_ENDPOINT
```

## 3. Send a sample trace

Creates a parent agent span and a nested tool span, then flushes to ingest.

```python
import time

import veritrix
from veritrix.config import get_config

veritrix.init(
    api_key=VERITRIX_API_KEY,
    endpoint=VERITRIX_ENDPOINT,
    default_tags=["getting-started", "colab"],
    framework="manual",
    agent_name="Colab Quickstart",
)

trace_id = get_config().trace_id
print(f"trace_id: {trace_id}")

with veritrix.trace(
    "hello-agent",
    span_type="agent",
    input_data={"message": "Hello from Google Colab"},
):
    time.sleep(0.1)
    with veritrix.trace("sample-tool", span_type="tool", input_data="ping"):
        time.sleep(0.05)

veritrix.end()
print("Done — open the Traces page in Veritrix to inspect this run.")
```
